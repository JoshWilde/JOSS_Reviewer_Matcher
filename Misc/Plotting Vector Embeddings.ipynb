```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
spaCy_vecs = np.load('JOSS_Paper_Vectors_SpaCY_TEMP.npy')
```

```python
spaCy_vecs.shape
```

```python
spaCy_vecs[:,1:].shape
```

```python
for i in range(10):
    X_embedded = TSNE(n_components=2).fit_transform(spaCy_vecs[:,1:])
    np.save('T-SNE_SpaCy_Embeddings_'+str(int(i)), X_embedded)
    plt.figure(figsize=(10,10))
    plt.scatter(X_embedded[:,0], X_embedded[:,1])
    plt.show()
```
